```python
# Step 3: Performance Score Calculation - CORRECTED
weights = {'CP':1, 'GT':1, 'C':3, 'DC':-3, 'ST':3, 'RO':3, 'MRO':-2, 'DH':2}

performance_data = {
    'Player Name': ['Rilee russouw','Phil Salt','Yash Dhull','Axer Patel','Lalit yadav','Aman Khan','Kuldeep yadav'],
    'CP': [2,1,3,2,1,4,3],      # 7 values
    'GT': [1,2,1,3,2,1,0],      # 7 values
    'C': [1,0,2,1,1,0,1],       # 7 values
    'DC': [0,1,0,0,0,0,1],      # 7 values
    'ST': [0,0,0],      # 7 values - fix chesa
    'RO': [0,1,0,0,0,1,0],      # 7 values
    'MRO': [0,0,1,0,0,0,0],     # 7 values
    'DH': [1,0,0,0,1,0,0],      # 7 values
    'RS': [2,-1,3,0,-2,1,4]     # 7 values
}

perf_df = pd.DataFrame(performance_data)

# PS Formula apply cheyadam
perf_df['PS'] = (perf_df['CP']*weights['CP'] + 
                 perf_df['GT']*weights['GT'] + 
                 perf_df['C']*weights['C'] + 
                 perf_df['DC']*weights['DC'] + 
                 perf_df['ST']*weights['ST'] + 
                 perf_df['RO']*weights['RO'] + 
                 perf_df['MRO']*weights['MRO'] + 
                 perf_df['DH']*weights['DH'] + 
                 perf_df['RS'])

print("ShadowFox Formula tho Performance Score Table:")
display(perf_df.sort_values('PS', ascending=False))
```

```python
# Step 3: Performance Score Calculation - FINAL
weights = {'CP':1, 'GT':1, 'C':3, 'DC':-3, 'ST':3, 'RO':3, 'MRO':-2, 'DH':2}

performance_data = {
    'Player Name': ['Rilee russouw','Phil Salt','Yash Dhull','Axer Patel','Lalit yadav','Aman Khan','Kuldeep yadav'],
    'CP': [2,1,3,2,1,4,3],
    'GT': [1,2,1,3,2,1,0],
    'C':  [1,0,2,1,1,0,1],
    'DC': [0,1,0,0,0,0,1],
    'ST': [0,0,0,0,0,0,0],
    'RO': [0,1,0,0,0,1,0],
    'MRO': [0,0,1,0,0,0,0],
    'DH': [1,0,0,0,1,0,0],
    'RS': [2,-1,3,0,-2,1,4]
}

perf_df = pd.DataFrame(performance_data)

# PS Formula - Idhi add cheyali
perf_df['PS'] = (perf_df['CP']*1 + perf_df['GT']*1 + perf_df['C']*3 + perf_df['DC']*-3 + 
                 perf_df['ST']*3 + perf_df['RO']*3 + perf_df['MRO']*-2 + perf_df['DH']*2 + perf_df['RS'])

print("ShadowFox Formula tho Performance Score Table:")
print(perf_df.sort_values('PS', ascending=False))
```

```python
import matplotlib.pyplot as plt

# Bar Graph - Top Players
top_players = perf_df.sort_values('PS', ascending=False).head(5)

plt.figure(figsize=(10,6))
plt.bar(top_players['Player Name'], top_players['PS'], color=['gold','silver','#CD7F32','skyblue','lightgreen'])
plt.title('ShadowFox Performance Score - Top 5 Players', fontsize=16, weight='bold')
plt.xlabel('Players')
plt.ylabel('Performance Score (PS)')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Top 5 players PS % ela divide ayyindi
plt.figure(figsize=(8,8))
plt.pie(top_players['PS'], labels=top_players['Player Name'], autopct='%1.1f%%', 
        colors=['gold','silver','#CD7F32','skyblue','lightgreen'], startangle=90)
plt.title('Top 5 Players - Performance Score Share', fontsize=16, weight='bold')
plt.show()
```
